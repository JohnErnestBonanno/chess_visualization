```python
def to_coordinate(move_list_algebraic):
    move_list_split = move_list_algebraic.split(' ') # split it 
    coordinate_notation = []
    
    #define starting position + board
    starting_fen = 'rnbqkbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBQKBNR w KQkq - 0 1'
    board = chess.Board(starting_fen)
    
    for moves in move_list_split:        
        if(moves == 'O-O'):
            piece = 'O-O'
        if(moves == 'O-O-O'):
            piece = 'O-O-O'
        if(moves[0] in ('K','Q','B','N','R')):
            piece = moves[0]
        if(moves[0] not in ('K','Q','B','N','R','O')):
            piece = 'p'
        new_move = board.push_san(moves).uci()
        new_move = piece + '_' + str(new_move)
        coordinate_notation.append(new_move)
    return(coordinate_notation) 
```

```python
coordinate_list = []
for x in df.moves:
    moves_in_coordinate = to_coordinate(x)
    coordinate_list.append(moves_in_coordinate)
```

```python
def distance_calc(coordinate_moves):

#intances of a pawn getting promoted, means it ends in q, so would need to trim it    
    
    white_distance = {'King':0,'Queen':0,'Rook':0,'Bishop':0,'Knight':0,'Pawn':0}
    black_distance = {'King':0,'Queen':0,'Rook':0,'Bishop':0,'Knight':0,'Pawn':0}

    for count, value in enumerate(coordinate_moves):
        if count % 2 == 0:
            if value[:1] == 'K':
                white_distance['King'] += 1
            if value[:1] == 'N':
                white_distance['Knight'] += 3
            if value[:1] == 'p':
                
                # new code 
                if value[-1] in ['q','r','n','b']:
                    value = value[:-1]
                    
                pawn_distance = abs(int(value[-1:]) - int(value[-3:-2]))
                white_distance['Pawn'] += pawn_distance
            if value[:1] == 'B':
                bishop_distance = abs(int(value[-1:]) - int(value[-3:-2]))
                white_distance['Bishop'] += bishop_distance
            if value[:1] == 'R':
                if abs(int(value[-3:-2]) - int(value[-1:])) != 0:
                    white_distance['Rook'] += abs(int(value[-3:-2]) - int(value[-1:]))
                else:
                    white_distance['Rook'] += abs(file_dict[value[-2:-1]]-file_dict[value[-4:-3]])
            if value[:1] == 'Q':
                if abs(int(value[-3:-2]) - int(value[-1:])) != 0:
                    white_distance['Queen'] += abs(int(value[-3:-2]) - int(value[-1:]))
                else:
                    white_distance['Queen'] += abs(file_dict[value[-2:-1]]-file_dict[value[-4:-3]])
            if 'O-O' in value:
                white_distance['King'] += 2
                white_distance['Rook'] += 2
            if 'O-O-O' in value:
                white_distance['King'] += 2 
                white_distance['Rook'] += 3
        if count % 2 == 1:
            if value[:1] == 'K':
                black_distance['King'] += 1
            if value[:1] == 'N':
                black_distance['Knight'] += 3
            if value[:1] == 'p':
               
            
                if value[-1] in ['q','r','n','b']:
                    value = value[:-1]
                
                pawn_distance = abs(int(value[-1:]) - int(value[-3:-2]))
                black_distance['Pawn'] += pawn_distance
            if value[:1] == 'B':
                bishop_distance = abs(int(value[-1:]) - int(value[-3:-2]))
                black_distance['Bishop'] += bishop_distance
            if value[:1] == 'R':
                if abs(int(value[-3:-2]) - int(value[-1:])) != 0:
                    black_distance['Rook'] += abs(int(value[-3:-2]) - int(value[-1:]))
                else:
                    black_distance['Rook'] += abs(file_dict[value[-2:-1]]-file_dict[value[-4:-3]])
            if value[:1] == 'Q':
                if abs(int(value[-3:-2]) - int(value[-1:])) != 0:
                    black_distance['Queen'] += abs(int(value[-3:-2]) - int(value[-1:]))
                else:
                    black_distance['Queen'] += abs(file_dict[value[-2:-1]]-file_dict[value[-4:-3]])
            if 'O-O-O' in value:
                black_distance['King'] += 2 
                black_distance['Rook'] += 3
            elif 'O-O' in value:
                black_distance['King'] += 2
                black_distance['Rook'] += 2
    
    white = pd.DataFrame(white_distance,index=[0])
    white.columns = white.columns = ['king_w','queen_w','rooks_w','bishops_w','knights_w','pawns_w']
    
    black = pd.DataFrame(black_distance,index=[0])
    black.columns = black.columns = ['king_b','queen_b','rooks_b','bishops_b','knights_b','pawns_b']
    
    test = pd.concat([white,black],axis='columns')
    
    return(test)
    
    #return(pd.DataFrame(white_distance,index=[0]))
           
           #pd.DataFrame.from_dict(black_distance,index=[0]))
    #return(white_distance.values(),black_distance.values())
```
