```python
#functions for pulling info - white pieces

#white
def king_w_calc(moves):
    king_w['move_list'].append(moves)
    if 'x' in moves:
        king_w['capture_count']+=1
    # kings can't deliver check, checkmate, or have ambiguous chess notaiton

def queens_w_calc(moves):
    queens_w['move_list'].append(moves)
    if 'x' in moves:
        queens_w['capture_count']+=1
    if '+' in moves:
        queens_w['check_count']+=1
    if '#' in moves:
        queens_w['checkmate_count'] = +1
    #currently working! 
    if sum(element in [char for char in moves] for element in file_list) >=2:
        queens_w['ambiguity_count'] +=1

def rooks_w_calc(moves):
    rooks_w['move_list'].append(moves)
    if 'x' in moves:
        rooks_w['capture_count']+=1
    if '+' in moves:
        rooks_w['check_count']+=1
    if '#' in moves:
        rooks_w['checkmate_count'] = +1
    # under original conditions, bishops don't have ambiguity because dark and light never cross
    #however, if player obtains new bishops via promotion, could have ambiguity
    if sum(element in [char for char in moves] for element in file_list) >=2:
        rooks_w['ambiguity_count'] +=1         
        
def bishops_w_calc(moves):
    bishops_w['move_list'].append(moves)
    if 'x' in moves:
        bishops_w['capture_count']+=1
    if '+' in moves:
        bishops_w['check_count']+=1
    if '#' in moves:
        bishops_w['checkmate_count'] = +1
    # under original conditions, bishops don't have ambiguity because dark and light never cross
    #however, if player obtains new bishops via promotion, could have ambiguity
    if sum(element in [char for char in moves] for element in file_list) >=2:
        bishops_w['ambiguity_count'] +=1

def knights_w_calc(moves):
    knights_w['move_list'].append(moves)
    if 'x' in moves:
        knights_w['capture_count']+=1
    if '+' in moves:
        knights_w['check_count']+=1
    if '#' in moves:
        knights_w['checkmate_count'] = +1
    # under original conditions, bishops don't have ambiguity because dark and light never cross
    #however, if player obtains new bishops via promotion, could have ambiguity
    if sum(element in [char for char in moves] for element in file_list) >=2:
        knights_w['ambiguity_count'] +=1 

def pawns_w_calc(moves):
    pawns_w['move_list'].append(moves)
    if 'x' in moves:
        pawns_w['capture_count']+=1
    if '+' in moves:
        pawns_w['check_count']+=1
    if '#' in moves:
        pawns_w['checkmate_count'] = +1
    # under original conditions, bishops don't have ambiguity because dark and light never cross
    #however, if player obtains new bishops via promotion, could have ambiguity
    if sum(element in [char for char in moves] for element in file_list) >=2:
        pawns_w['ambiguity_count'] +=1 
             
def castle_w_calc(moves):
    castle_w['move_list'].append(moves)
    if '+' in moves:
        castle_w['check_count']+=1
    if '#' in moves:
        castle_w['checkmate_count'] = +1
    # castling cannot:
        # capture a piece
        # be ambiguous
```

```python
def game_calc(i):
    game_id = df.id[i]
    moves = df.moves[i]
    
    moves = moves.split(" ")
    
    white_moves = moves[::2]
    black_moves = moves[1::2]
    
    for moves in white_moves:
        if 'K' in moves:
            king_w_calc(moves)
        if 'Q' in moves:
            queens_w_calc(moves)
        if 'B' in moves:
            bishops_w_calc(moves)
        if 'N' in moves:
            knights_w_calc(moves)
        if 'R' in moves:
            rooks_w_calc(moves)
        if ('O-O' in moves or 'O-O-O' in moves): #fun note, castling is defined by 'O' not 0
            castle_w_calc(moves)
        if moves[0] not in ['K','Q','B','N','R'] and moves !='O-O' and moves != 'O-O-O':
            pawns_w_calc(moves)
    
    capture_w = [king_w['capture_count'],
                 queens_w['capture_count'],
                 rooks_w['capture_count'],
                 bishops_w['capture_count'],
                 knights_w['capture_count'],
                 pawns_w['capture_count']]
    
    check_w =   [queens_w['check_count'],
                 rooks_w['check_count'],
                 bishops_w['check_count'],
                 knights_w['check_count'],
                 pawns_w['check_count']]
    
    checkmate_w = [queens_w['checkmate_count'],
                 rooks_w['checkmate_count'],
                 bishops_w['checkmate_count'],
                 knights_w['checkmate_count'],
                 pawns_w['checkmate_count']]
   
    for moves in black_moves:
        if 'K' in moves:
            king_b_calc(moves)
        if 'Q' in moves:
            queens_b_calc(moves)
        if 'B' in moves:
            bishops_b_calc(moves)
        if 'N' in moves:
            knights_b_calc(moves)
        if 'R' in moves:
            rooks_b_calc(moves)
        if ('O-O' in moves or 'O-O-O' in moves): #fun note, castling is defined by 'O' not 0
            castle_b_calc(moves)
        if moves[0] not in ['K','Q','B','N','R'] and moves !='O-O' and moves != 'O-O-O':
            pawns_b_calc(moves)
    
    capture_b = [king_b['capture_count'],
                 queens_b['capture_count'],
                 rooks_b['capture_count'],
                 bishops_b['capture_count'],
                 knights_b['capture_count'],
                 pawns_b['capture_count']]
    
    check_b = [queens_b['check_count'],
                 rooks_b['check_count'],
                 bishops_b['check_count'],
                 knights_b['check_count'],
                 pawns_b['check_count']]
    
    checkmate_b = [queens_b['checkmate_count'],
                 rooks_b['checkmate_count'],
                 bishops_b['checkmate_count'],
                 knights_b['checkmate_count'],
                 pawns_b['checkmate_count']]
    
    temp_df_capture = pd.DataFrame([game_id] + capture_w + capture_b)
    temp_df_capture = temp_df_capture.T
    temp_df_capture.columns = ['game_id','king_w','queen_w','rooks_w','bishops_w','knights_w','pawns_w','king_b','queen_b','rooks_b','bishops_b','knights_b','pawns_b']
    
    temp_df_check = pd.DataFrame([game_id] + check_w + check_b)
    temp_df_check = temp_df_check.T
    temp_df_check.columns = ['game_id','queen_w','rooks_w','bishops_w','knights_w','pawns_w','queen_b','rooks_b','bishops_b','knights_b','pawns_b']
    
    temp_df_checkmate = pd.DataFrame([game_id] + checkmate_w + checkmate_b)
    temp_df_checkmate = temp_df_checkmate.T
    temp_df_checkmate.columns = ['game_id','queen_w','rooks_w','bishops_w','knights_w','pawns_w','queen_b','rooks_b','bishops_b','knights_b','pawns_b']
    
    
    return((temp_df_capture, temp_df_check, temp_df_checkmate))
```
